```python
#Loading Libraries

import nltk
from nltk import word_tokenize,sent_tokenize
import nltk.data
from __future__ import division
from urllib import urlopen
from bs4 import BeautifulSoup
from nltk import pos_tag
```

```python
#url = "http://news.bbc.co.uk/2/hi/health/2284783.stm"
#url = "https://www.garyvaynerchuk.com/two-secrets-staying-motivated-hungry/"
from urllib import urlopen
url = "http://www.bbc.com/news/health-38312995"
raw = urlopen(url).read()
print raw
```

```python
## http://stackoverflow.com/questions/22799990/beatifulsoup4-get-text-still-has-javascript
## Cleaning HTML Tags
## The NLTK has its own cleanHTML function, but the Beautifulsoup package is more efficient
soup = BeautifulSoup(raw, 'html.parser')
##soup.get_text()

for script in soup(["script", "style"]):
    script.extract()    
text = soup.get_text()

# break into lines and remove leading and trailing space on each
lines = (line.strip() for line in text.splitlines())
# break multi-headlines into a line each
chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
# drop blank lines
text = '\n'.join(chunk for chunk in chunks if chunk)

print text;
```

```python
# Sentence Segmentation
sent_tokenizer=nltk.data.load('tokenizers/punkt/english.pickle')
sentences1 = nltk.sent_tokenize(text)
sentences1 [1:10]
```

```python
#Word Tokenization
tokens = [nltk.word_tokenize(s) for s in sentences1]
tokens[1]
```

```python
#Part of Speech Tagging
pos_tag = [nltk.pos_tag(t) for t in tokens]
pos_tag[1:10]
```

```python
#Chunking
#Noun Phrase Chunking
grammar = "NP: {<DT>?<JJ>*<NN>}"
cp = nltk.RegexpParser(grammar)

result = [cp.parse(word) for word in pos_tag]
print result
result[0].draw()
```

```python
result[0].draw()
```

```python
# chunk determiner/possessive, adjectives and noun
# chunk sequences of proper nouns
grammar1 = r"""
  NP: {<DT|PP\$>?<JJ>*<NN>}   
      {<NNP>+}                
"""
cp1 = nltk.RegexpParser(grammar1)
result1 = [cp1.parse(word) for word in pos_tag]
print result1
```
